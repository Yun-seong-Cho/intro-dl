```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### What is numpy and matplotlib? 

#### A.numpy란 Numerical Python의 줄임말이며 파이썬의 고성능 수치계산을 위한 라이브러리이다. 여러 형태의 벡터나 행렬 계산을 가능토록 한다. matplotlib이란 다양한 데이터를 많은 방법으로 도식화 할 수 있도록 하는 파이썬 라이브러리로 데이터를 시각화하기 좋은 방법이다.

```python

# Let's make a linear function
def make_linear(w=0.5, b=0.8, size=50, noise=1.0): 
    x = np.random.rand(size) 
    y = w * x + b 
    noise = np.random.uniform(-abs(noise), abs(noise), size=y.shape)
    yy = y + noise 
    plt.figure(figsize=(10, 7)) 
    plt.plot(x, y, color='r', label=f'y = {w}*x + {b}')
    plt.scatter(x, yy, label='data') 
    plt.legend(fontsize=10)
    plt.show()
    print(f'w: {w}, b: {b}')
    return x, yy 
```

```python
x, y = make_linear(w=0.3, b=0.5, size=20, noise=0.01)

y[5] = 0.75
y[10] = 0.75

plt.figure(figsize=(10, 7))
plt.scatter(x, y)
plt.show()
```

```python
w = np.random.uniform(low=-1.0, high=1.0)
b = np.random.uniform(low=-1.0, high=1.0)
```

```python
y_hat = w * x + b

error = (y_hat - y) ** 2
```

### Quick Question

#### (1) Write a name of the loss function above. HInt: _ _ _ Loss

#### (2) Is it okay to use 'error = (y_hat-y)' or 'error = np.abs(y_hat-y)'?
 #### Write your own answer and justify your choice.

#### Your Answer


#### (1)MSE loss function 
#### (2)Yes. If you want to consider both the direction and magnitude of errors, choose 'error = (y_hat-y). (this resulting error can be positive or negative, indicating whether the presiction is overestimating or underestimating the true values). And choose 'error = np.abs(y_hat-y)' if you are mainly interested inthe magnitude of errors and want a simpler metric.

```python
num_epoch = 5000 #epoch 설정

learning_rate = 0.5 # Learning rate 설정

## Question 1
errors = []
w = np.random.uniform(low=-1.0, high=1.0) #w,b 랜덤설정
b = np.random.uniform(low=-1.0, high=1.0)

for epoch in range(num_epoch): #0~4999번 실행
    y_hat = x * w + b #y_hat 설정

    error = ((y_hat - y)**2).mean() #loss function 설정
    if error < 0.0005:  
        break

    w = w - learning_rate * ((y_hat - y) * x).mean() #w,b 업데이트
    b = b - learning_rate * (y_hat - y).mean()

    errors.append(error)

    if epoch % 5 == 0: #에폭 5마다 과정 프린트하기
        print("{0:2} w = {1:.5f}, b = {2:.5f} error = {3:.5f}".format(epoch, w, b, error))

print("----" * 15) #결과 도출
print("{0:2} w = {1:.1f}, b = {2:.1f} error = {3:.5f}".format(epoch, w, b, error))
```
