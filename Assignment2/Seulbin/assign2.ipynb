```python
def make_linear(w=0.5, b=0.8, size=50, noise=1.0):
    x = np.random.rand(size)
    y = w * x + b
    noise = np.random.uniform(-abs(noise), abs(noise), size=y.shape)
    yy = y + noise
    plt.figure(figsize=(10, 7))
    plt.plot(x, y, color='r', label=f'y = {w}*x + {b}')
    plt.scatter(x, yy, label='data')
    plt.legend(fontsize=20)
    plt.show()
    print(f'w: {w}, b:{b}')
    return x, yy
```

```python
x, y = make_linear(w=0.3, b=0.5, size=100, noise=0.01)

y[5] = 0.75
y[10] = 0.75

plt.figure(figsize=(10, 7))
plt.scatter(x, y)
plt.show()
```

```python
w = np.random.uniform(low=-1.0, high=1.0)
b = np.random.uniform(low=-1.0, high=1.0)
```

```python
y_hat = w * x + b

error = (y_hat - y) ** 2
```

```python
##(1) Mean Squared Error(MSE) Loss
##(2) 'error = (y_hat - y)'를 사용하는 것 보다 MSE loss function을 사용하는 것이 일반적이다.
## MSE를 사용하면 제곱을 하기 때문에 상대적인 크기를 갖게 되어 스케일의 영향을 줄일 수 있다. Loss function을 최적화하는 과정에서 미분이 중요한데 제곱 오차를 사용하면 미분이 간단하게 이루어진다.
##'error = np.abs(y_hat - y)'(MAE)는 Absolute Error를 사용하는 방법으로 outlier의 영향을 덜 받고 예측값과 실제값 간의 크기 차이를 직접 나타내므로 해석이 쉽지만, 이상치가 무시될 수 있고 미분이 불가능하다는 단점이 있다.
```

```python
num_epoch = 5000

learning_rate = 0.5
```

```python
##(1)Epoch refers to the one entire passing of training data through the algorithm.
##(2)Learning rate is a tuning parameter in an optimization algorithm that determines the step size at each iteration while moving toward a minimum of a loss function.
```

Question1

```python
errors = [] #에러값을 저장할 빈 리스트를 생성
w = np.random.uniform(low=-1.0, high=1.0) #가중치 w를 -1.0에서 1.0 사이의 무작위 값으로 초기화
b = np.random.uniform(low=-1.0, high=1.0) #편향 b를 -1.0에서 1.0 사이의 무작위 값으로 초기화

for epoch in range(num_epoch):
    y_hat = x * w + b #현재의 가중치와 편향을 사용하여 예측값 계산

    error = ((y_hat - y)**2).mean() #평균 제곱 오차 계산
    if error < 0.0005:
        break #만약 평균 제곱 오차가 0.0005보다 작다면 훈련 종료

    #경사 하강법을 사용하여 가중치와 편향을 업데이트
    w = w - learning_rate * ((y_hat - y) * x).mean()
    b = b - learning_rate * (y_hat - y).mean()

    errors.append(error) #현재 에러값을 리스트에 추가

    if epoch % 5 == 0:
        print("{0:2} w = {1:.5f}, b = {2:.5f} error = {3:.5f}". format(epoch, w, b, error)) #epoch이 5의 배수일 때, 현재의 epoch, 가중치, 편향, 에러값을 출력

#최종 결과 출력
print("----" * 15)
print("{0:2} w = {1:.1f}, b = {2:.1f} error = {3:.5f}".format(epoch, w, b, error))
```
