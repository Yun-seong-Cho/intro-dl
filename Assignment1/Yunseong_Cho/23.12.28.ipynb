## Simple MLP

```python
# 1. Module Import

import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms, datasets
```

```python
if torch.cuda.is_available():
    DEVICE = torch.device('cuda')
else:
    DEVICE = torch.device('cpu')

print('Using PyTorch version:', torch.__version__, ' Device:', DEVICE)
# 딥러닝 모델을 학습과 실행을 GPU 환경에서 할지 CPU환경에서 할지 선택.
```

```python
BATCH_SIZE = 32
EPOCHS = 10
# 하이퍼파라미터 설정(배치의 사이즈와 학습 반복 횟수)
```

```python
# 3. MNIST Download (Train set, Test set split)

train_dataset = datasets.MNIST(root = "../data/MNIST",
                               train = True,
                               download = True,
                               transform = transforms.ToTensor())
# pytorch에서 기본적으로 제공하는 MNIST 데이터셋을 root로 다운로드, 학습데이터셋을 생성(true)하고 데이터셋이 없다면 인터넷에서 다운로드(true)하며 이미지를 tensor로 변환)
test_dataset = datasets.MNIST(root = "../data/MNIST",
                              train = False,
                              transform = transforms.ToTensor())
# 테스트 데이터셋에 대해서는 학습을 진행하지 않음(false)
train_loader = torch.utils.data.DataLoader(dataset = train_dataset,
                                           batch_size = BATCH_SIZE,
                                           shuffle = True)
# 다운받은 데이터들을 미니배치로 나누고 셔플합니다.
test_loader = torch.utils.data.DataLoader(dataset = test_dataset,
                                          batch_size = BATCH_SIZE,
                                          shuffle = False)
# 테스트 데이터들에 대해서는 셔플을 진행하지 않습니다.
```

```python
for (X_train, y_train) in train_loader:
    print('X_train:', X_train.size(), 'type:', X_train.type())
    print('y_train:', y_train.size(), 'type:', y_train.type())
    break
# 하나의 미니배치에 존재하는 32개의 이미지-라벨 쌍의 사이즈와 타입을 프린트
```

```python
pltsize = 1
plt.figure(figsize=(10 * pltsize, pltsize))
for i in range(10):
    plt.subplot(1, 10, i + 1)
    plt.axis('off')
    plt.imshow(X_train[i, :, :, :].numpy().reshape(28, 28), cmap = "gray_r")
    plt.title('Class: ' + str(y_train[i].item()))

# 미니배치에서 앞에서 10개의 학습 데이터를 그림으로 그리고 그에 따른 라벨도 표시.
```

```python
# Multi Layer Perceptron

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(28 * 28, 512) # 784 -> 512 개의 노드 수를 가지는 fully connected layer
        self.fc2 = nn.Linear(512, 256) # 512 -> 256 개의 노드 수를 가지는 fully connected layer
        self.fc3 = nn.Linear(256, 10) # 256 -> 10 개의 노드 수를 가지는 fully connected layer

    def forward(self, x):
        x = x.view(-1, 28 * 28) # 2D인 이미지 텐서를 flatten하여 1D 텐서로 변환
        x = self.fc1(x) # fully connected layer
        x = F.sigmoid(x) # Activation function으로 sigmoid function
        x = self.fc2(x) # fully connected layer
        x = F.sigmoid(x) # Activation function으로 sigmoid function
        x = self.fc3(x) # fully connected layer
        x = F.log_softmax(x, dim = 1) # 소프트맥스보다 기울기 소실이 적은 로그소프트맥스 사용.
        return x
```

```python
# Optimizer, Objective Function

model = Net().to(DEVICE)
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01, momentum = 0.5)
criterion = nn.CrossEntropyLoss()

print(model)
# 설정한 디바이스로 네트워크(모델)을 생성하고, optimizer로 SGD(learning rate는 0.01, momentum은 0.5)로 설정하고 손실함수는 다중 클래스 분류에 많이 사용되는 cross entropy loss를 사용.
```

```python
def train(model, train_loader, optimizer, log_interval):
    model.train() # 모델을 학습 상태로 설정
    for batch_idx, (image, label) in enumerate(train_loader): # 배치에서 몇번째, 이미지텐서, 라벨을 iterate
        image = image.to(DEVICE)
        label = label.to(DEVICE)
        optimizer.zero_grad() # 그래디언트 초기화
        output = model(image) # 모델에 이미지 텐서를 입력
        loss = criterion(output, label) # 입력값으로부터 나온 output과 라벨을 통해서 loss 계산
        loss.backward() # backpropagation 실행
        optimizer.step() # 파라미터 최적화

        if batch_idx % log_interval == 0: # log_interval 간격으로 아래의 학습 진행도를 프린트
            print("Train Epoch: {} [{}/{} ({:.0f}%)]\tTrain Loss: {:.6f}".format(epoch, batch_idx * len(image),len(train_loader.dataset), 100. * batch_idx / len(train_loader),loss.item()))
# 기존에 만든 모델 네트워크의 파라미터들을 학습 데이터들로 학습하는 과정.
```

```python
def evaluate(model, test_loader):
    model.eval() # 모델을 테스트 상태로 설정
    test_loss = 0
    correct = 0 # 평가 기준 초기화

    with torch.no_grad(): # 그래디언트 계산 비활성화
        for image, label in test_loader: # 학습때와 마찬가지로 테스트 데이터에 있는 이미지 텐서와 라벨 iterate
            image = image.to(DEVICE)
            label = label.to(DEVICE)
            output = model(image) # 테스트 데이터에 존재하는 이미지 텐서를 모델에 입력
            test_loss += criterion(output, label).item() # 각 미니배치에 대한 테스트 loss를 누적
            prediction = output.max(1, keepdim = True)[1] # output 텐서에서 가장 큰 값을 가지는 인덱스를 출력
            correct += prediction.eq(label.view_as(prediction)).sum().item() # 각 미니배치에 대한 테스트 정확도를 누적

    test_loss /= (len(test_loader.dataset) / BATCH_SIZE) # 테스트 데이터셋에 대한 평균 loss 계산
    test_accuracy = 100. * correct / len(test_loader.dataset) # 테스트 데이터셋에 대한 평균 정확도 계산
    return test_loss, test_accuracy
```

```python
for epoch in range(1, EPOCHS + 1): # 10번의 epoch을 진행
    train(model, train_loader, optimizer, log_interval = 200) # 각 조건을 만족하게 훈련 진행
    test_loss, test_accuracy = evaluate(model, test_loader) # evaluate 함수를 통해서 학습한 모델을 평가하여 손실과 정확도 계산
    print("\n[EPOCH: {}], \tTest Loss: {:.4f}, \tTest Accuracy: {:.2f} % \n".format(epoch, test_loss, test_accuracy)) # 학습 진행과 손실 정확도를 출력
```
