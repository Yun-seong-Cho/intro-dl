## Simple MLP

*   Reference. https://github.com/Justin-A/DeepLearning101/blob/master/2-1_MNIST_MLP.ipynb

```python
''' 1. Module Import '''
import numpy as np                            # import numpy and call it as np
import matplotlib.pyplot as plt               # import pyplot from matplotlib and call it as plt

import torch                                  # import torch
import torch.nn as nn                         # import nn from torch and call it as nn
import torch.nn.functional as F               # import functional from nn and call it F
from torchvision import transforms, datasets  # import transforms and datasets from tochvision
```

```python
if torch.cuda.is_available():             # Check if a CUDA-enabled GPU is available
  DEVICE = torch.device('cuda')           # If a GPU is available, assign the device 'cuda' to the variable 'DEVICE'
else:                                     # If a GPU is not available, the code executes the following block.
  DEVICE = torch.device('cpu')            # If a GPU is not availabe, the 'cpu' device is assigned to the variable 'DEVICE'

print('Using PyTorch version:', torch.__version__, ' Device:', DEVICE)    # Print the Pytorch version and the selected device
```

```python
BATCH_SIZE = 32       # The number of training examples in one iteration
EPOCHS = 10           # size of epoch, the number of the model will be trained on the entire dataset.
```

```python
''' 3. MNIST Download (Train set, Test set split) '''
train_dataset = datasets.MNIST(root = "../data/MNIST",
                               train = True,
                               download = True,
                               transform = transforms.ToTensor())       # A training dataset

test_dataset = datasets.MNIST(root = "../data/MNIST",
                              train = False,
                              transform = transforms.ToTensor())        # A test dataset

train_loader = torch.utils.data.DataLoader(dataset = train_dataset,
                                           batch_size = BATCH_SIZE,
                                           shuffle = True)              # A data loader for the training dataset

test_loader = torch.utils.data.DataLoader(dataset = test_dataset,
                                          batch_size = BATCH_SIZE,
                                          shuffle = False)              # A data loader for the test dataset
```

```python
## Quick Question
# What is MNIST?
# Google it and write the answer

# Answer
# Modified National Institute of Standards and Technology - A large databese of handwritten digits (0 through 9).
```

```python
for (X_train, Y_train) in train_loader:                                # Start a loop
  print('X_train:', X_train.size(), 'type:', X_train.type())           # Print the input data information
  print('Y_train:', Y_train.size(), 'type:', Y_train.type())           # Print the target label information
  break                                                                # Exit the loof after processing the first batch.
```

```python
pltsize = 1                                                 # Initialize the variable with 1
plt.figure(figsize = (10 * pltsize, pltsize))               # Create a new Matplotlib figure
for i in range(10):                                         # Start a loop that iterates 10 times.
  plt.subplot(1, 10, i + 1)                                 # Creat a subplot in the figure
  plt.axis('off')                                           # Turn off zxis labels
  plt.imshow(X_train[i, :, :, :].numpy().reshape(28, 28), cmap = "gray_r")      # Display an image on the current subplot
  plt.title('Class: ' + str(Y_train[i].item()))             # Set the title of the current subplot
```

```python
''' Multi Layer Perceptron '''
class Net(nn.Module):                     # Define a new class named 'Net' that inherits from 'nn.Module'
  def __init__(self):                     # Begin the definition of the class constructor
    super(Net, self).__init__()           # Call the constructor of the parent class ('nn.Module')
    self.fc1 = nn.Linear(28 * 28, 512)    # Define a fully connected (dense) layer using 'nn.Linear' (Input size of 28*28 and Output size of 512)
    self.fc2 = nn.Linear(512, 256)        # Define a fully connected (dense) layer using 'nn.Linear' (Input size of 512 and Output size of 256)
    self.fc3 = nn.Linear(256, 10)         # Define a fully connected (dense) layer using 'nn.Linear' (Input size of 256 and Output size of 10)

  def forward(self, x):               # Begin the definition of the forward pass method
    x = x.view(-1, 28 * 28)           # Reshape the input tensor 'x' into a 1D tensor with a size of 28*28.
    x = self.fc1(x)                   # Apply the first fully connected layer('fc1') followed by a sigmoid activation function
    x = F.sigmoid(x)
    x = self.fc2(x)                   # Apply the second fully connected layer('fc2') followed by a sigmoid activation function
    x = F.sigmoid(x)
    x = self.fc3(x)                   # Apply the third fully connected layer('fc3') followed by a log softmax activation along dimension 1.
    x = F.log_softmax(x, dim = 1)
    return x                          # Return the final output tensor
```

```python
''' Optimizer, Objective Function '''
model = Net().to(DEVICE)                    # Create an instance of the 'Net' class and move it to 'DEVICE'
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01, momentum = 0.5)          # Create a SGD(Stochastic Gradient Descent) optimizer with the parameter 'model', the learning rate 0.01, and the momentum 0.5.
criterion = nn.CrossEntropyLoss()           # Define the loss criterion 'nn.CrossEntropyLoss()' combines the softmax activation with the negative log likelihood loss.

print(model)                                # Print the architecture and configuration of the neural network model.
```

```python
def train(model, train_loader, optimizer, log_interval):          # Define a model 'train' with four parameters: 'model'(neural network model), 'train_loader'(training data loader), 'optimizer'(training optimizer), and 'log_interval'(interval for logging information).
  model.train()                                                   # Set the neural network model to training mode.
  for batch_idx, (image, label) in enumerate(train_loader):       # A loop over batches in the 'train_loader'; The index of the current batch 'batch_idx';
                                                                  # '(image, label)' contains the input images and corresponding labels.
      image = image.to(DEVICE)                                    # Move the input images to the device 'DEVICE'
      label = label.to(DEVICE)                                    # Move the input image labels to the device 'DEVICE'
      optimizer.zero_grad()                                       # Zero the gradients of the model parameters.
      output = model(image)                                       # Pass the input images through the model.
      loss = criterion(output, label)                             # Calculates the loss by comparing the model predictions ('output') to the ground truth labels ('label')
      loss.backward()                                             # Perform gackpropagation
      optimizer.step()                                            # Update the model parameters using the optimizer's update rule.

      if batch_idx % log_interval == 0:                           # Prints training pregress information.
        print("Train Epoch: {} [{}/{} ({:.0f}%)]\tTrain Loss: {:.6f}".format(epoch, batch_idx * len(image),
                                                                             len(train_loader.dataset), 100. * batch_idx / len(train_loader), loss.item()))
```

```python
def evaluate(model, test_loader):         # Defind a function 'evaluate' and 'test_loader'
  model.eval()                            # Set the neural network model to evaluation
  test_loss = 0                           # Initialize the total test loss as 0
  correct = 0                             # Initialize the number of correct predictions as 0

  with torch.no_grad():                                                     # Open a context where gradient computation is disabled.
    for image, label in test_loader:                                        # Start a loop over batches in the 'test_loader'; The input images 'image' and the corresponding labels 'label'
        image = image.to(DEVICE)                                            # Move the input images to the device 'DEVICE'
        label = label.to(DEVICE)                                            # Move the input image labels to the device 'DEVICE'
        output = model(image)                                               # Get the output predictions by the model from the input images
        test_loss += criterion(output, label).item()                        # Calculate the test loss by comparing the 'output' and the 'label' using the loss criterion.
        prediction = output.max(1, keepdim = True)[1]                       # Get the index of the maximum value along dimension [1]
        correct += prediction.eq(label.view_as(prediction)).sum().item()    # Update the count of correct predictions; Compare 'prediction' and 'label'

  test_loss /= (len(test_loader.dataset) / BATCH_SIZE)                      # Normalize the test loss by dividing it by the ratio of the total number of samples in the test dataset to the batch size.
  test_accuracy = 100. * correct / len(test_loader.dataset)                 # Calculate the test accuracy as. a percentage
  return test_loss, test_accuracy                                           # Return the normalized test loss and test accuracy
```

```python
for epoch in range(1, EPOCHS + 1):                                        # Start a loop that iterates over 'EPOCHS'
    train(model, train_loader, optimizer, log_interval = 200)             # Call the 'train' function to perform training
    test_loss, test_accuracy = evaluate(model, test_loader)               # Call the 'evaluate' fucntion to assess the model's performance on the dataset.
    print("\n[EPOCH: {}], \tTest Loss: {:.4f}, \tTest Accuracy: {:.2f} % \n".format(epoch, test_loss, test_accuracy))     #Print a summary of the test results after each epoch.
```
